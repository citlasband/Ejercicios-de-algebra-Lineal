## Parte 2: Distancias y KNN

```python
from scipy.spatial.distance import euclidean, cityblock
```

### Producto escalar con `dot()` y operador `@`

**¿Qué se busca?** Calcular el producto escalar entre dos vectores.

**¿Cómo se hace?** Usamos `np.dot()` o el operador `@` entre dos arrays NumPy.

**¿Para qué sirve?** Es una operación básica en álgebra lineal y modelos ML.


**Conclusión:** El producto escalar mide la similitud en dirección entre dos vectores.

```python

v1 = np.array([1, 2])
v2 = np.array([3, 4])
print("Producto escalar con dot:", np.dot(v1, v2))
print("Producto escalar con @:", v1 @ v2)
```

### Distancia euclidiana y Manhattan

**¿Qué se busca?** Calcular diferentes distancias entre vectores.

**¿Cómo se hace?** Usamos `euclidean()` y `cityblock()`.

**¿Para qué sirve?** Para medir similitud o diferencia entre puntos.


**Conclusión:** Elegir entre distancias depende del contexto: Euclidiana para continuidad, Manhattan para cuadrículas.

```python

print("Distancia Euclidiana:", euclidean(v1, v2))
print("Distancia Manhattan:", cityblock(v1, v2))
```

### Comparación de distancias con vectores multidimensionales

**¿Qué se busca?** Comparar distancias entre vectores en espacios mayores a 2D.

**¿Cómo se hace?** Se aplican las funciones en vectores de más dimensiones.

**¿Para qué sirve?** Para análisis de similitud en datasets complejos.


**Conclusión:** Las métricas se escalan naturalmente a cualquier número de dimensiones.

```python

a = np.array([2, 3, 5])
b = np.array([5, 1, 7])
print("Euclidiana:", euclidean(a, b))
print("Manhattan:", cityblock(a, b))
```

### Implementar KNN (1-NN) simple con Euclidiana

**¿Qué se busca?** Clasificar un punto basado en su vecino más cercano.

**¿Cómo se hace?** Se calcula la distancia entre el punto consulta y los existentes.

**¿Para qué sirve?** KNN es un modelo simple pero poderoso para clasificación.


**Conclusión:** KNN utiliza distancias para asignar clases sin entrenamiento complejo.

```python

def knn(query, data, labels):
    distances = [euclidean(query, x) for x in data]
    min_idx = np.argmin(distances)
    return labels[min_idx]

data = np.array([[1,2],[3,4],[5,6]])
labels = ['A', 'B', 'C']
knn(np.array([4,5]), data, labels)
```

### Crear clase básica de modelo ML

**¿Qué se busca?** Crear una clase de modelo simple.

**¿Cómo se hace?** Definiendo `fit` y `predict` en una clase Python.

**¿Para qué sirve?** Para sentar las bases de modelos personalizables.


**Conclusión:** Podemos construir estructuras básicas de modelos desde cero para entender su funcionamiento interno.

```python

class ModeloDummy:
    def fit(self, X, y):
        self.mean = np.mean(X, axis=0)

    def predict(self, X):
        return [self.mean for _ in X]

modelo = ModeloDummy()
modelo.fit(data, labels)
modelo.predict([[0,0]])
```

### Suma, resta, multiplicación de matrices

**¿Qué se busca?** Realizar operaciones algebraicas entre matrices.

**¿Para qué sirve?** Para aplicar modelos lineales y operaciones matemáticas.
